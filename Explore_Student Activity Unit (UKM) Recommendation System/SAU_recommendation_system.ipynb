```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
UKM = pd.read_csv('ukm_dataset.csv')
```

```python
UKM.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ukm_features = UKM.drop(columns=["ID_UKM", "Nama_UKM"])
```

```python
scaler = MinMaxScaler()
ukm_features_scaled = scaler.fit_transform(ukm_features)
```

```python
ukm_features_scaled_df = pd.DataFrame(ukm_features_scaled, columns=ukm_features.columns)
```

```python
ukm_features_scaled_df.head()
```

```python
husein = pd.DataFrame([{
    "Seni": 0.1,
    "Sains": 0.8,
    "Sosial": 0.4,
    "Riset": 0.8,
    "Olahraga": 0.2,
    "Kerohanian": 0.3,
    "Kompetitif": 0.8,
    "Waktu_Komitmen": 0.5,
    "Ekstrovert_Friendly": 0.7
}])
```

```python
husein
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(ukm_features_scaled_df, husein)
rekomendasi_df = UKM.copy()
rekomendasi_df["Similarity"] = similarities
```

```python
top_n = rekomendasi_df.sort_values(by="Similarity", ascending=False).head(5)
```
